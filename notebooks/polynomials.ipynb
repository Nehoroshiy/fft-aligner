```python
def inverse_1_2(small_matrix, dtype=np.float64):
    """Performs inverse of small matrices of size 1 or 2 by simple formulas

    Args:
        small_matrix (np.array): matrix for inverse searching

    Returns:
        np.array: inverse of small matrix

    Raises:
        Exception: An error occurred because matrix is not 1x1 or 2x2
        Exception: An error occurred because matrix of size 1x1 or 2x2 is singular
    """
    if small_matrix.shape == (1, 1):
        if small_matrix[0, 0] == 0:
            raise Exception('Matrix of size 1x1 is singular')
        return np.array([[1.0 / small_matrix[0, 0]]], dtype=dtype)
    if small_matrix.shape != (2, 2):
        raise Exception("Matrix isn't 2x2 matrix")
    a = small_matrix[0, 0]
    b = small_matrix[0, 1]
    c = small_matrix[1, 0]
    d = small_matrix[1, 1]
    det = a * d - b * c
    if det == 0:
        raise Exception('Matrix of size 2x2 is singular')
    inverse = np.zeros(small_matrix.shape, dtype=dtype)
    inverse[0, 0] = d / det
    inverse[0, 1] = -c / det
    inverse[1, 0] = -b / det
    inverse[1, 1] = a / det
    return inverse
```

```python
def recursive_inversion(A):
    n = A.shape[0]
    if n <= 2:
        return inverse_1_2(A)
    k = n / 2
    B = A[:k, :k]
    C = A[k:, :k]
    D = A[k:, k:]
    A_INV = np.zeros([n, n])
    B_INV = recursive_inversion(B)
    W = dot(C, B_INV)
    X = dot(W, C.T)
    S = D - X
    A_INV[k:, k:] = recursive_inversion(S)
    V = A_INV[k:, k:]
    A_INV[k:, :k] = -dot(V, W)
    A_INV[:k, k:] = A_INV[k:, :k].T
    Z = dot(W.T, -A_INV[k:, :k])
    A_INV[:k, :k] = B_INV + Z
    return A_INV
```

```python
def simple_evaluation(polynomial, x):
    n = polynomial.size
    powers = np.array([x ** i for i in xrange(n)])
    return dot(polynomial, powers)
```

```python
def res_evaluation(polynomial, x):
    res = polynomial[-1]
    n = polynomial.size
    for i in xrange(n-2, -1, -1):
        res = x * res + polynomial[i]
    return res
```

```python
def rev_polynom(polynom):
    if len(polynom.shape) == 1:
        # we have coefficient representation of a polynom, then simply reverse coefficients
        return polynom[::-1]
    else:
        if len(polynom.shape) == 2 and polynom.shape[0] == 2:
            n = polynom.shape[1]
            # we have point-value representation
            power_matrix = np.array([[x ** i for i in xrange(n)] for x in polynom[0]])
            coefficients_reversed = np.linalg.solve(power_matrix, polynom[1])[::-1]
            return np.array([polynom[0], dot(power_matrix, coefficients_reversed)])
        else:
            raise Exception('unrecognized type of input. Input must be 1-d vector or [2, n] matrix')
```

```python
#returns smallest power of 2 that is greater or equal than v
def power_2_bound(v):
    v -= 1
    v |= v >> 1
    v |= v >> 2
    v |= v >> 4
    v |= v >> 8
    v |= v >> 16
    return v + 1
```

```python
def fft_fast(a):
    N = a.shape[0]
    
    if N & (N - 1) != 0:
        raise ValueError("size of x must be a power of 2")
    
    N_min = min(N, 32)
    
    n = np.arange(N_min)
    k = n[:, None]
    M = np.exp(-2j * np.pi * n * k / N_min)
    X = np.dot(M, a.reshape([N_min, -1]))
    
    n = np.arange(N_min)
    while X.shape[0] < N:
        X_even = X[:, :X.shape[1] / 2]
        X_odd = X[:, X.shape[1] / 2:]
        factor = np.exp(-1j * np.pi * np.arange(X.shape[0])
                        / X.shape[0])[:, None]
        X = np.vstack([X_even + factor * X_odd,
                       X_even - factor * X_odd])
    return X.ravel()

def ifft_fast(y):
    N = y.shape[0]
    
    if N & (N - 1) != 0:
        raise ValueError("size of x must be a power of 2")
    
    N_min = min(N, 32)
    
    n = np.arange(N_min)
    k = n[:, None]
    #T = np.exp(-2j * np.pi * n * k / N_min)
    #M = np.conj(T / N_min)
    M = np.exp((2j * np.pi * n) * k / N_min) / N_min
    X = np.dot(M, y.reshape([N_min, -1]))
    
    n = np.arange(N_min)
    while X.shape[0] < N:
        X_even = X[:, :X.shape[1] / 2]
        X_odd = X[:, X.shape[1] / 2:]
        factor = np.exp(1j * np.pi * np.arange(X.shape[0])
                        / X.shape[0])[:, None]
        X = np.vstack([X_even + factor * X_odd,
                       X_even - factor * X_odd])
    return X.ravel()
```
