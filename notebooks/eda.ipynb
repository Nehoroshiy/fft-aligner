```python
import pandas as pd
import numpy as np
import random
import seaborn as sns
import pylab
```

```python
df = pd.read_csv('data.csv', delimiter=',').values
random.shuffle(df)
columns = pd.read_csv('data.csv', delimiter=',').columns[:-1]
```

```python
X, Y = np.array([np.nan_to_num(np.array(x[:-1], dtype=np.float)) for x in df]), np.array(map(lambda x: 1 if x == 'group_two' else -1, [y[-1] for y in df]))
```

### Логистичиская реграссия

#### F-мера, на кросс валидации из 5 разбиений

```python
from sklearn import linear_model, cross_validation, metrics
clf = linear_model.SGDClassifier(loss='log')
print 'значение F-меры на кросс валидации = %s' % (sum(cross_validation.cross_val_score(clf, X, Y, cv=5, scoring='f1')) / 5)
```

#### ROC кривая, размер обучающей выборки 250, тестовой 132

```python
clf = linear_model.SGDClassifier(loss='log')
clf.fit(X[:250], Y[:250])
pY = clf.predict_proba(X[250:])
fpr, tpr, thresholds = metrics.roc_curve(Y[250:], pY[:, 1])

# Plot ROC curve
roc_auc = metrics.auc(fpr, tpr)
pylab.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % roc_auc)
pylab.plot([0, 1], [0, 1], 'k--')
pylab.xlim([0.0, 1.0])
pylab.ylim([0.0, 1.0])
pylab.xlabel('False Positive Rate')
pylab.ylabel('True Positive Rate')
pylab.title('ROC')
```

#### Веса признаков: Отрицательные веса - положительное влияние, положительные веса --- отрицательное влияние. Если вес по модулю не большой, то его влияние нельзя назвать достоверным.

```python
pd.DataFrame(sorted(zip(columns, clf.coef_[0]), key=lambda x: x[1]))
```

```python
for r in sorted(zip(columns, clf.coef_[0]), key=lambda x: x[1]):
    print r[0], r[1]
```

### Дерево решений

```python
df = pd.read_csv('data.csv', delimiter=',')
columns = pd.read_csv('data.csv', delimiter=',').columns[:-1]
for i, t in enumerate(filter(lambda x: 'dlitel_nost__funkzionirovania' in x, columns)):
    X = df[t]
    df[t] = [i if x == 1 else 0 for x in X]
df['dlitel_nost__funkzionirovania_poslednej_plomby'] = df['dlitel_nost__funkzionirovania_poslednej_plomby_menee_1_goda']+df['dlitel_nost__funkzionirovania_poslednej_plomby_1_to_3_let']+df['dlitel_nost__funkzionirovania_poslednej_plomby_3_to_5_let'] + df['dlitel_nost__funkzionirovania_poslednej_plomby_5_to_7_let'] + df['dlitel_nost__funkzionirovania_poslednej_plomby_7_to_10_let']+ df['dlitel_nost__funkzionirovania_poslednej_plomby_bolee_10_let']
df = df.drop('dlitel_nost__funkzionirovania_poslednej_plomby_menee_1_goda', 1)
df = df.drop('dlitel_nost__funkzionirovania_poslednej_plomby_3_to_5_let', 1)
df = df.drop('dlitel_nost__funkzionirovania_poslednej_plomby_5_to_7_let', 1)
df = df.drop('dlitel_nost__funkzionirovania_poslednej_plomby_1_to_3_let', 1)
df = df.drop('dlitel_nost__funkzionirovania_poslednej_plomby_7_to_10_let', 1)
df = df.drop('dlitel_nost__funkzionirovania_poslednej_plomby_bolee_10_let', 1)

x = df['group']
df=df.drop('group', 1)
df['group'] = x

columns = df.columns[:-1]
df = df.values
random.shuffle(df)

X, Y = np.array([np.nan_to_num(np.array(x[:-1], dtype=np.float)) for x in df]), np.array(map(lambda x: 1 if x == 'group_two' else -1, [y[-1] for y in df]))
```

#### F-мера, на кросс валидации из 5 разбиений

```python
from sklearn import tree
clf = tree.DecisionTreeClassifier(min_samples_split=16, min_samples_leaf=7, criterion='entropy')
print 'значение F-меры на кросс валидации = %s' % (sum(cross_validation.cross_val_score(clf, X, Y, cv=5, scoring='f1')) / 5)
```

Мы можем наблюдать значение f-меры, больше чем у линеной модели, это показывает, что дерево решений гораздо лучше описывает выборку.

#### ROC кривая, размер обучающей выборки 300, тестовой 82

```python
clf = tree.DecisionTreeClassifier(min_samples_split=16, min_samples_leaf=7, criterion='entropy')
clf.fit(X[:300], Y[:300])
pY = clf.predict_proba(X[300:])
fpr, tpr, thresholds = metrics.roc_curve(Y[300:], pY[:, 1])

# Plot ROC curve
roc_auc = metrics.auc(fpr, tpr)
pylab.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % roc_auc)
pylab.plot([0, 1], [0, 1], 'k--')
pylab.xlim([0.0, 1.0])
pylab.ylim([0.0, 1.0])
pylab.xlabel('False Positive Rate')
pylab.ylabel('True Positive Rate')
pylab.title('ROC')
```

### Спроецируем признаки в трех пространство с помошью методо главных компонент.

```python
df = pd.read_csv('data.csv', delimiter=',')
columns = pd.read_csv('data.csv', delimiter=',').columns[:-1]
for i, t in enumerate(filter(lambda x: 'dlitel_nost__funkzionirovania' in x, columns)):
    X = df[t]
    df[t] = [i if x == 1 else 0 for x in X]
df['dlitel_nost__funkzionirovania_poslednej_plomby'] = df['dlitel_nost__funkzionirovania_poslednej_plomby_menee_1_goda']+df['dlitel_nost__funkzionirovania_poslednej_plomby_1_to_3_let']+df['dlitel_nost__funkzionirovania_poslednej_plomby_3_to_5_let'] + df['dlitel_nost__funkzionirovania_poslednej_plomby_5_to_7_let'] + df['dlitel_nost__funkzionirovania_poslednej_plomby_7_to_10_let']+ df['dlitel_nost__funkzionirovania_poslednej_plomby_bolee_10_let']
df = df.drop('dlitel_nost__funkzionirovania_poslednej_plomby_menee_1_goda', 1)
df = df.drop('dlitel_nost__funkzionirovania_poslednej_plomby_3_to_5_let', 1)
df = df.drop('dlitel_nost__funkzionirovania_poslednej_plomby_5_to_7_let', 1)
df = df.drop('dlitel_nost__funkzionirovania_poslednej_plomby_1_to_3_let', 1)
df = df.drop('dlitel_nost__funkzionirovania_poslednej_plomby_7_to_10_let', 1)
df = df.drop('dlitel_nost__funkzionirovania_poslednej_plomby_bolee_10_let', 1)

x = df['group']
df=df.drop('group', 1)
df['group'] = x
df = df.values
```

```python
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D

pca = PCA(n_components=3)

Xp, Yp = np.array([np.nan_to_num(np.array(x[:-1], dtype=np.float)) for x in df]), np.array(map(lambda x: 1 if x == 'group_two' else -1, [y[-1] for y in df]))

pca.fit(Xp)
Xpca = pca.transform(Xp)
fig = pylab.figure()
ax = fig.add_subplot(111, projection='3d')
n = Yp.size

Xt = Xpca.T
red_flag = Yp == -1
blue_flag = Yp != -1
ax.scatter(Xt[0, Yp == -1], Xt[1, Yp == -1], Xt[2, Yp == -1], c='r', marker='o')
ax.scatter(Xt[0, blue_flag], Xt[1, blue_flag], Xt[2, blue_flag], c='b', marker='^')
```
