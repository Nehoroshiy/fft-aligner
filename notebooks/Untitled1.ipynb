```python
def get_points_on_circle(n=5, r=1.0):
    root_args = np.array([1.0*(i*2*math.pi)/n for i in xrange(n)])
    roots = np.exp(1j * root_args)
    return roots
```

```python
circle = get_points_on_circle(200)
```

```python
bbox_props = dict(boxstyle="round,pad=0.3", fc="white", ec="b", lw=1, alpha=0.7)

f = sns.plt.figure(figsize=[10, 10])
sns.plt.plot(roots.real, roots.imag, 'o')
sns.plt.plot(circle.real, circle.imag, '--')
for i in xrange(-1, n-1, 1):
    sns.plt.plot(roots[i:i+2].real, roots[i:i+2].imag, '--r', alpha=0.7)
sns.plt.plot(roots[::roots.size - 1].real, roots[::roots.size - 1].imag, '--r', alpha=0.7)
#sns.plt.plot(roots[0:2].real, roots[0:2].imag, '--r', alpha=0.7)
for i in xrange(n):
    sns.plt.plot(np.array([0, roots[i].real]), np.array([0, roots[i].imag]), '--b', alpha=0.2)

sns.plt.title(r"complex roots of unity of power 5")
sns.plt.xlabel(r"$Re z$")
sns.plt.ylabel(r"$Im z$")

sns.plt.annotate(r'$\omega_n^0$', xy=(roots[0].real, roots[0].imag), bbox=bbox_props)
sns.plt.annotate(r'$\omega_n^1$', xy=(roots[1].real, roots[1].imag), xytext=(roots[1].real, roots[1].imag - 0.05), bbox=bbox_props)
sns.plt.annotate(r'$\omega_n^2$', xy=(roots[2].real, roots[2].imag), bbox=bbox_props)
sns.plt.annotate(r'$\omega_n^3$', xy=(roots[3].real, roots[3].imag), bbox=bbox_props)
sns.plt.annotate(r'$\omega_n^4$', xy=(roots[4].real, roots[4].imag), bbox=bbox_props)
            #arrowprops=dict(facecolor='black', shrink=0.05),
sns.plt.subplots_adjust(left=0.15, right=0.95, top=0.9, bottom=0.1)
sns.plt.savefig('unity_roots.pdf', dpi=100)
```
